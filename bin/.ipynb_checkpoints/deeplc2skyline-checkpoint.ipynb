```python
# read in the peak boundaries that are currently only for one file
import pandas as pd
import sys
peak_boundaries_ex = pd.read_csv('../results/peak boundaries.csv')
keepers = ['20191125_QEHFX_lkp_histone-DIA_A_05000_043.raw', 
           '20191125_QEHFX_lkp_histone-DIA_A_05000_080.raw', 
           '20191125_QEHFX_lkp_histone-DIA_A_05000_117.raw']
peak_boundaries_ex = peak_boundaries_ex.loc[peak_boundaries_ex['File Name'].isin(keepers)]
peak_boundaries_ex.drop(columns=['File Name', 'Precursor Charge', 'PrecursorIsDecoy'], inplace=True)
peak_boundaries_ex = peak_boundaries_ex.groupby('Peptide Modified Sequence').mean().reset_index()
peak_boundaries_ex.head()


# grab all the files (I guess just from the directory?)
# fuck it I'm lazy
file_names = pd.read_csv('../results/file_names.csv')
file_names.head()

# "copy" the peak boundary information for all the files
peak_boundaries = pd.DataFrame()
for file in list(file_names['Replicate']):
    sys.stdout.write(file+'\r')
    temp = peak_boundaries_ex.copy()
    temp['File Name'] = file
    peak_boundaries = peak_boundaries.append(temp)

peak_boundaries.fillna(str('NaN'), inplace=True)

# write out the csv for a skyline import
peak_boundaries.to_csv('../results/peak boundaries fixed.csv', index=False)
peak_boundaries.tail()
```
